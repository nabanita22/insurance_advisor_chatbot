```python
import fitz  # PyMuPDF
from pathlib import Path

pdf_path = Path("../data/raw_brochures/Star_Health_Assure.pdf")

doc = fitz.open(pdf_path)

print("PDF loaded successfully")
print("Total pages:", len(doc))
```

```python
page = doc[0]
text = page.get_text()

print(text[:1000])
```

```python
pages_text = []

for page_num in range(len(doc)):
    page = doc[page_num]
    text = page.get_text()

    pages_text.append({
        "page_number": page_num + 1,
        "text": text
    })

print("Total pages extracted:", len(pages_text))
```

```python
# View first 2 pages
for p in pages_text[:2]:
    print("="*60)
    print(f"PAGE {p['page_number']}")
    print("="*60)
    print(p["text"][:800])
```

```python
full_text = "\n".join([p["text"] for p in pages_text])

print("Total characters in document:", len(full_text))
```

```python
output_path = "../data/processed_products/raw_text_star_health_assure.txt"

with open(output_path, "w", encoding="utf-8") as f:
    f.write(full_text)

print("Raw text saved successfully")
```
